```python
import numpy as np 
import pandas as pd 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
var = data.var()
print("Variance : \n",var)
```

```python
cov = data.cov()
print("Covariance : \n",cov)    
```

```python
corr = data.corr()
print("Correlation : \n",corr)
```
